```python
def count_pixels(image):
    return image.shape[0] * image.shape[1] * image.shape[2]

def count_labels(labels):
    # don't count the background label
    return len(da.unique(labels.data).compute())-1
```

```python
for name, sdata in [
    ("Tonsil_2022", harpy.datasets.macsima_tonsil()), 
    ("Liu_2022", harpy.datasets.pixie_example()),
    ("IMMUcan_2021", harpy.datasets.proteomics.imc_example()),
]:
    print(f"Dataset: {name}")
    count = 0
    pixel_counts = []
    labels_counts = []
    for idx, image_name in enumerate([x for x in sdata.images if 'nucle' not in x]):
        print(f"  Image {idx}: {image_name}")
        # calc total number of pixels
        image = sdata[image_name]
        pixel_count = count_pixels(image)
        pixel_counts.append(pixel_count)
        count += 1
        print(f"  Number of pixels: {pixel_count}")

    # count unique cells in labels
    for idx, labels_name in enumerate([x for x in sdata.labels if 'nucle' not in x]):
        print(f"  Labels {idx}: {labels_name}")
        # calc total number of pixels
        labels = count_labels(sdata[labels_name])
        print(f"  Number of labels: {labels}")
        labels_counts.append(labels)
    print(f"  Number of images: {count}")
    #  Avg. number of pixels in all images as 10^9
    print(f"  Avg. number of pixels in all images: {sum(pixel_counts) / len(pixel_counts) / 1e9:.3f} billion")
    #  Avg. number of labels in all labels
    if labels_counts:
        print(f"  Avg. number of labels in all images: {sum(labels_counts) / len(labels_counts)}")
```
