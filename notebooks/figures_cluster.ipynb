## Wall clock time

```python
hue_order = ['flowsom_batch', 'flowsom', 'pyflowsom']
df['method'] = pd.Categorical(df['method'], hue_order)
# remove non-batch methods with threads more than 1 thread
df_subset = df[
    (df['method'] == 'flowsom_batch') |
    (df['method'] == 'flowsom') & (df['threads'] == 1) |
    (df['method'] == 'pyflowsom') & (df['threads'] == 1)
]
df_subset
```

```python
# set colors to orange, blue, purple and green
cmap = sns.color_palette("Dark2")
# change order in cmap, switch first two colors
cmap[0], cmap[1] = cmap[1], cmap[0]
sns.set_palette(cmap)

dash_styles = {
    1: (1, 0),
    2: (4, 1),
    4: (2, 1),
    8: (1, 1),
}

axs = sns.lineplot(data=df_subset, x="pixels", y="wall_clock_time", hue="method", marker='o', style='threads', err_style="bars", dashes=dash_styles)
plt.xlabel('Dataset size (pixels)')
plt.ylabel('Wall clock time (s)')

# make logarithmic
plt.yscale('log', base=10)
plt.xscale('log', base=10)
# # add more ticks to y
# plt.yticks([100, 1000, 10_000], ['100', '1000', '10000'])
# import matplotlib as mpl
# axs.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
# plt.grid(True, which="both")
postprocess(axs)
```

```python
# plot number of pixels and wall clock time on the same plot for each method and each thread count
axs = sns.lineplot(data=df, x="pixels", y="wall_clock_time", hue="method", marker='o', err_style="bars", style='threads', dashes=dash_styles)
postprocess(axs)
plt.xlabel('Dataset size (pixels)')
plt.ylabel('Wall clock time (s)')
plt.xscale('log', base=10)
plt.yscale('log', base=10)
```
